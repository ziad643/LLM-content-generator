```python
# Define our system prompt - you can experiment with this later, changing the last sentence to 'Respond in markdown in Spanish."

system_prompt = """
You are a financial market analysis assistant specialized in cryptocurrency markets.

You are allowed to access live website data to retrieve the current Bitcoin (BTC) price and today’s market details, including:
- Current price
- 24h price change (percentage and absolute)
- Trading volume
- Market trend (bullish, bearish, or sideways)
- Notable intraday movements or volatility

Your task is to:
1. Clearly summarize today's Bitcoin price and key market metrics in simple, student-friendly language.
2. Analyze the short-term market sentiment based ONLY on the observed data (price movement, volume, volatility).
3. Provide a reasoned *buy / hold / sell* opinion for *short-term trading*, explaining:
   - Why this action might be reasonable today
   - What signals support this view
4. Explicitly mention potential risks and uncertainty.

Guidelines:
- Avoid hype or emotional language.
- Be neutral, analytical, and cautious.

Output format:
- Today's Bitcoin Snapshot
- Market Analysis
- Suggested Action (Buy / Hold / Sell)
- Reasoning
- Risks & Disclaimer
"""
```

```python
# Define our user prompt

user_prompt_prefix = """
Access a reliable cryptocurrency price website and retrieve today's Bitcoin (BTC) market data.

Based on today's data:
- Summarize the current Bitcoin price and 24-hour movement
- Analyze the short-term market sentiment
- Recommend whether to buy, hold, or sell Bitcoin today
- Clearly explain the reasoning behind your recommendation
- Mention key risks and uncertainties

Present the result in a clear, structured format suitable for a student or beginner
"""
```

```python
# See how this function creates exactly the format above

def messages_for(website):
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt_prefix + website}
    ]
```

```python
# And now: call the OpenAI API. You will get very familiar with this!

def summarize(url):
    website = fetch_website_contents(url)
    response = gemini.chat.completions.create(
        model = "gemini-2.5-flash",
        messages = messages_for(website)
    )
    return response.choices[0].message.content
```

```python
# A function to display this nicely in the output, using markdown
from IPython.display import display, Markdown
def display_summary(url):
    summary = summarize(url)
    display(Markdown(summary))
```

```python
display_summary("https://coinmarketcap.com/currencies/bitcoin/")
```
